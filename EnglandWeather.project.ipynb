```python
new = Data["Formatted Date"].str.split(" ", n = 2 , expand =True)
new
```

```python
Data1 = Data.drop(columns =["Formatted Date"])
Data1
```

```python
Data1.insert (0 , "Formatted Date" , new[0],True)
```

```python
Data2 =Data1[::24]
Data2
```

```python
Data2.sort_values(by = "Formatted Date" ,inplace =True)
Data2
```

```python
#میانگین برای دمای هوا:
```

```python
Df1 = Data1.sort_values(by="Formatted Date")
DataMean = Df1.groupby(pd.Grouper(key="Formatted Date"), as_index=False).mean()
Df2 = DataMean
Df2
```

```python
Df3 = Df2.dropna()
Df3
```

```python
Df3.describe()
```

```python
import seaborn as sns
features = ["Temperature (C)", "Wind Speed (km/h)", "Pressure (millibars)"]
target = "Humidity"
sns.pairplot(data=Df3, vars=features, hue=target, palette='viridis')
plt.show()
```

```python
x = Df3["Temperature (C)"]
y = Df3["Humidity"]
plt.scatter(x, y, c='blue', alpha=0.2) 
plt.xlabel("Temperature (C)")
plt.ylabel("Humidity")
plt.title("H VS T")
plt.grid()
plt.show()
```

```python
x = Df3["Wind Speed (km/h)"]
y = Df3["Humidity"]
plt.scatter(x, y, c='blue', alpha=0.2) 
plt.xlabel("Wind Speed (km/h)")
plt.ylabel("Humidity")
plt.title("H VS W")
plt.grid()
plt.show()
```

```python
x = Df3["Pressure (millibars)"]
y = Df3["Humidity"]
plt.scatter(x, y, c='blue', alpha=0.2) 
plt.xlabel("Pressure (millibars)")
plt.ylabel("Humidity")
plt.title("H VS P")
plt.grid()
plt.show()
```
